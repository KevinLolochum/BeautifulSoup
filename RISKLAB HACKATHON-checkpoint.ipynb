DATA CLEANING

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
%matplotlib inline
```

```python
#load the data
R = pd.read_csv('data/Risklab.csv', delimiter = ',')
R.head(3)
```

```python
R.shape
```

```python
# duplicated values
R.duplicated().sum()
```

```python
# deleting duplicated values and keeping only the first
R = R.drop_duplicates(keep = 'first')
R.duplicated().sum()
```

```python
R.shape
```

```python
#checking for null values
R.isnull().sum()
```

```python
#dropping all null columns
R1 = R.dropna(how = 'any', axis = 0, thresh = None, subset = None, inplace = False)
R1.shape
```

```python
#cehcking datatypes
R1.dtypes
```

Dealing with outliers and unknown values using a boolen vector.

```python
R1 = R1[((R1['Credit_Limit'] >= 200.0) & (R1['Credit_Limit'] <= 40000.0 )) & 
       ((R1['Sex'] >= 1.0) & (R1['Sex'] <= 2.0)) & 
       ((R1['Education'] >= 1.0) & (R1['Education'] <= 4.0 )) & 
       ((R1['Marital_Status'] >= 1.0) & (R1['Marital_Status'] <= 3.0 ))&
       (R1['Age'] <= 100.0 )
        
       ]


R1.describe()
```

```python
#reset index to keep the data clean
R1.reset_index(drop = True)
R1.head()
```

```python
R2 = R1
```

```python
# I will change sex to string, male and female instead of a float. 
# I will also change Marital status and education to categorical
R2['Sex'] = R2['Sex'].astype('int')
R2['Sex'].dtypes
```
